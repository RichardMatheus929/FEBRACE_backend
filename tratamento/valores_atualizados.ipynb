Criando duas listas de dicionarios para os dados brutos e os atualmente 

```python
import pandas as pd
import requests

# criar os brutos,tem que transformar os valores de anos para string
df_bruto = pd.read_csv('checkpoints/checkpoint(0).csv')
df_bruto = df_bruto.sort_values(by='id')
dados_brutos = df_bruto.to_dict(orient='records')
for proj in dados_brutos:
    proj['ano'] = str(proj['ano'])


# criar os tratados
allprojects = requests.get("http://localhost:8000/projetos/").json()
df_tratado = pd.DataFrame(allprojects)
df_tratado = df_tratado.sort_values(by='id')
allprojects = df_tratado.to_dict(orient='records')

allprojects[0],dados_brutos[0]
```

```python
# Converter os valores em um conjunto de strings para poder fazer a diferença entre eles
allprojects = set([str(proj) for proj in allprojects])
dados_brutos = set([str(proj) for proj in dados_brutos])

len(allprojects),len(dados_brutos)
```

df_atualizado contém apenas os valores que agora estão no banco de dados, de uma forma diferente do que estavam antes

```python
dados_atualizados = list(allprojects.difference(dados_brutos))
dados_atualizados = [eval(proj) for proj in dados_atualizados]

df_atualizado = pd.DataFrame(dados_atualizados).sort_values(by='id')
df_atualizado
```
