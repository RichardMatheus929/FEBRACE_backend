```python
import pandas as pd
import requests

allprojects = requests.get("http://localhost:8000/projetos/").json()
df_DATABASE = pd.DataFrame(allprojects)
df = pd.DataFrame(allprojects)
```

```python
def atualizar_projeto(projeto):

    import requests
    from datetime import datetime

    data_hora_atual = datetime.now().strftime('%d-%m-%Y %H:%M:%S')
    projeto_modificado = requests.get(f"http://localhost:8000/projetos/{projeto['id']}/").json()
    projeto_novo = requests.put(f"http://localhost:8000/projetos/{projeto['id']}/",json=projeto).json()

    with open('log_alterações.txt','a',encoding='utf-8') as log:

        antigo = f"Valor antigo: {str(projeto_modificado)} \n"
        novo = f"Atualizado : {str(projeto_novo)} \n"

        log.write(antigo)
        log.write(novo)
        log.write(str(data_hora_atual)+" \n")
        log.write('\n')

def criar_projeto(projeto):
    import requests
    projeto = requests.post('http://localhost:8000/projetos/',json=projeto).json()
    print(projeto)
```
